This notebook generates sets of simulated mock HST quality lens systems with added mass complexity to the plane of the main deflector.

```python
#Initiating autoreload
%load_ext autoreload
%autoreload 2
%matplotlib inline

#Adding parent directory to working path
import os
import sys

parent_dir = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
sys.path.append(parent_dir)

#Improting packages
import numpy as np
from complexLA import lens_parameters, test_set_generation
```

### 1) Generate Sample Parameters ###

First we will need to generate a catalog of lens parameters that are drawn from the distributions in the csv file we are calling here.

If you already have the parameters for your test sets then this module is optional.

User defined parameters:
- 'sample_num' - How many lenses will be generated

```python
#Defining how many lenses will be generated in the sample
sample_num = 101
```

```python
param_names = ['index','z_lens', 'gamma_md', 'theta_E_md', 'e1_md', 'e2_md', 'center_x_md', 'center_y_md', 'gamma1_md', 'gamma2_md', 'p_center_x', 'p_center_y', 'z_source', 'mag_app_source', 'R_sersic_source',
               'n_sersic_source', 'e1_source', 'e2_source', 'center_x_source', 'center_y_source', 'z_lens_light', 'mag_app_light', 'R_sersic_light', 'n_sersic_light', 'e1_light', 'e2_light', 
               'z_point_source', 'x_point_source', 'y_point_source', 'mag_app_point_source', 'sigma_sub']

#Generating the parameters to be used in the sample
param_dict = lens_parameters.perturberparameters(sample_num)
```

```python
with open('../docs/perturber_parameters_catalog.csv', 'w+') as f:
    np.savetxt(f, param_names, fmt='%s', newline=',')
    f.write('\n')
    np.savetxt(f, param_dict, fmt='%1.15f', delimiter=',')
```

### 2) Generate Test Sets ###
Now we will generate test sets of lenses.

User defined paramters:
- 'test_sets' - Array of test sets to be generated
    
    Options:
    - 'main_deflector' - Single main deflector without added complex mass
    - 'dark_perturber' - Adding a dark perturber to the main deflector
    - 'luminous_perturber' - Adding a luminous perturber to the main deflector
    - 'dark_substructure' - Adding dark subhalos to the main deflector

```python
#Declare array of test sets
test_sets = ['main_deflector', 'dark_perturber', 'luminous_perturber', 'dark_substructure']

#Generate simulated lenses and save images to an image folder
generation = test_set_generation.TestSetGeneration(sample_num, test_sets)
generation.generate_images()

#Display images
generation.display_sample()
```
